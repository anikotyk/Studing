\begin{align*}
-10.3x_1 + 10.2x_2 &= 70 \\
4.7x_1 + 12.3x_2 &= 173 \\
13.2x_1 + 8.8x_2 &= 282
\end{align*}

```python
import numpy as np
import matplotlib.pyplot as plt

# Визначення прямої
x = np.linspace(-10, 10, 100)

# Коефіцієнти прямої
a = -10.3
b = 10.2
c = 70

# Визначення y
y = (c - a * x) / b

# Побудова графіку
plt.figure(figsize=(8, 6))
plt.plot(x, y, linestyle='-', color='red', label=f'{a}x + {b}y = {c}')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Графік першої прямої')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Визначення прямих
x = np.linspace(-100, 100, 1000)

# Коефіцієнти прямих
a_values = [-10.3, 4.7, 13.2]
b_values = [10.2, 12.3, 8.8]
c_values = [70, 173, 282]

# Визначення y для кожної прямої
y_values = [(c - a * x) / b for a, b, c in zip(a_values, b_values, c_values)]

# Побудова графіку
plt.figure(figsize=(8, 16), dpi=100)
plt.grid(True, linestyle='--', color='gray')
plt.gca().set_facecolor('#f0f0f0')

# Побудова прямих
for i in range(3):
    linestyle = ['-', '--', '-.'][i]  # Визначення типу лінії
    color = plt.cm.tab10(i / 3)  # Визначення кольору
    label = f'${a_values[i]}x + {b_values[i]}y = {c_values[i]}$'  # Підпис
    plt.plot(x, y_values[i], linestyle=linestyle, color=color, label=label)

# Додавання підписів рівнянь прямих
for i in range(3):
    x_pos = 10
    y_pos = y_values[i][500]  # Вибір координати для підпису
    plt.text(x_pos, y_pos, f'${a_values[i]}x + {b_values[i]}y = {c_values[i]}$', rotation=-25)

# Додавання підписів осей та заголовку графіку
plt.xlabel('x')
plt.ylabel('y')
plt.title('Графік прямих та їх перетин')
plt.xlim(-100, 100)
plt.ylim(-100, 100)

# Додавання легенди
plt.legend()

# Заповнення кольором області, утвореної перетином всіх прямих
plt.fill_between(x, y_values[0], y_values[2], color='lightgray', alpha=0.5)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Коефіцієнти прямих
coefficients = np.array([
    [-10.3, 10.2, -70],
    [4.7, 12.3, -173],
    [13.2, 8.8, -282]
])

# Знаходження точки перетину прямих
intersection_point = np.linalg.lstsq(coefficients[:, :-1], coefficients[:, -1], rcond=None)[0]

# Визначення прямих
x = np.linspace(-100, 100, 1000)
y_values = [(coefficients[i, 2] - coefficients[i, 0] * x) / coefficients[i, 1] for i in range(3)]

# Побудова графіку
plt.figure(figsize=(8, 16), dpi=100)
plt.grid(True, linestyle='--', color='gray')
plt.gca().set_facecolor('#f0f0f0')

# Побудова прямих
for i in range(3):
    linestyle = ['-', '--', '-.'][i]  # Визначення типу лінії
    color = plt.cm.tab10(i / 3)  # Визначення кольору
    label = f'${coefficients[i, 0]}x + {coefficients[i, 1]}y = {coefficients[i, 2]}$'  # Підпис
    plt.plot(x, y_values[i], linestyle=linestyle, color=color, label=label)

# Додавання підписів рівнянь прямих
for i in range(3):
    x_pos = 10
    y_pos = y_values[i][500]  # Вибір координати для підпису
    plt.text(x_pos, y_pos, f'${coefficients[i, 0]}x + {coefficients[i, 1]}y = {coefficients[i, 2]}$', rotation=-25)

# Відмітка точки перетину
plt.scatter(intersection_point[0], intersection_point[1], color='red', marker='o')
plt.text(intersection_point[0], intersection_point[1], f'({intersection_point[0]:.2f}, {intersection_point[1]:.2f})', ha='right')

# Додавання підписів осей та заголовку графіку
plt.xlabel('x')
plt.ylabel('y')
plt.title('Графік прямих та їх перетин')
plt.xlim(-100, 100)
plt.ylim(-100, 100)

# Додавання легенди
plt.legend()

plt.show()
```
